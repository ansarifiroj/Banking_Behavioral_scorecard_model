# Getting outliars using violin plot

```python
plt.figure(figsize = (5, 5))
plt.title('Col6')
sns.violinplot(y = data.Col6)
```

# Calculating outliars

```python
Q1 = data.Col6.quantile(0.25)
Q3 = data.Col6.quantile(0.75)
IQR = Q3 - Q1
lower_limit = Q1 - 1.5*IQR
upper_limit = Q3 + 1.5*IQR
Data_outliers=data[(data.Col6<lower_limit)|(data.Col6>upper_limit)]
print(Data_outliers.shape)
data=data[(data.Col6>lower_limit)&(data.Col6<upper_limit)]#dropping the outliers from dataset
plt.figure(figsize = (5, 5))
plt.title('Col6')
sns.boxplot(y = data.Col6)
```

# Removing outliars

```python
Q1 = data.Col6.quantile(0.25)
Q3 = data.Col6.quantile(0.75)
IQR = Q3 - Q1
lower_limit = Q1 - 1.5*IQR
upper_limit = Q3 + 1.5*IQR
print(lower_limit,upper_limit)
mean_values=data['Col6'].mean()
data.loc[data.Col6 > 8.0, 'Col6'] = np.nan
data['Col6']=data.Col6.fillna(mean_values)
plt.figure(figsize = (5, 5))
plt.title('Col6')
sns.boxplot(y = data.Col6)
```
